```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib.colors
import os
from pylab import rcParams
#os.chdir('data')


# Read the data
data = pd.read_csv("Using Data/solar-energy-consumption-by-region.csv")
#ride_data = pd.read_csv("data/Cost_of_Residual_Fuel_Oil_Delivered_to_Electric_Generating_Plants_Annual.csv")

data.head()
```

```python
Middle_East = data[data.Country == 'Middle East']
Africa = data[data.Country == 'Africa']
South_Central_America = data[data.Country == 'South_Central_America']
Asia_Pacific = data[data.Country == 'Asia Pacific']
European_Union = data[data.Country == 'European Union']
North_America = data[data.Country == 'North America']
World = data[data.Country == 'World']
```

```python
#picked the regions only to be able to create the pie chart 
plt.plot(Middle_East.Year, Middle_East.Solar_PV_Terawatt_hours) 
plt.plot(Africa.Year, Africa.Solar_PV_Terawatt_hours) 
plt.plot(South_Central_America.Year, South_Central_America.Solar_PV_Terawatt_hours) 
plt.plot(Asia_Pacific.Year, Asia_Pacific.Solar_PV_Terawatt_hours) 
plt.plot(European_Union.Year, European_Union.Solar_PV_Terawatt_hours) 
plt.plot(North_America.Year, North_America.Solar_PV_Terawatt_hours)
key = plt.legend(labels = ['Middle East', 'Africa', 'South_Central_America',
                           'Asia Pacific','European Union','North America'], 
                 title= "Region", fontsize = 15)
rcParams['figure.figsize'] = 15, 15
key.legendHandles[1]._sizes = [60]
key.legendHandles[2]._sizes = [60]
key.legendHandles[3]._sizes = [60]
key.legendHandles[4]._sizes = [60]
key.legendHandles[5]._sizes = [60]
#key.legendHandles[6]._sizes = [60]
plt.ylabel('Measured in terawatt-hours')
plt.xlabel('Year')
plt.title('Solar PV energy consumption by region')
```

```python
group_names = ['Middle East', 'Africa', 'South America',
                           'Asia Pacific','European Union','North America']
```

```python
#As a percentage compared to the six regions 
# Pie chart
group_names = ['Middle 1%', 'Africa 1%', 'South America 1%',
                           'Asia Pacific 33%','European Union 49%','North America 15%']
sizes = [1, 1, 1, 33, 49, 15]
#colors
colors = a, b, c, d, e, f=[plt.cm.Blues, plt.cm.Reds, plt.cm.Greens, plt.cm.Greys, plt.cm.Purples, plt.cm.Oranges]

fig, ax = plt.subplots()
#ax.axis('equal')
mypie, _ = ax.pie(sizes, radius=2, labels=group_names, colors=[a(0.6), b(0.6), c(0.6), d(0.6), e(0.6), f(0.6)] )
plt.setp( mypie, width=0.2, edgecolor='white')
ax.set_title('Solar PV energy \n consumption by %')

plt.tight_layout()
# plt.show()
plt.savefig('Solar PV energy consumption by region.png',dpi=500)
```
