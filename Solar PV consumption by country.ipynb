```python
# %matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib.colors
import os
from pylab import rcParams
#os.chdir('data')

# Read the data
data = pd.read_csv("Using Data/solar-energy-consumption-by-country.csv")
#ride_data = pd.read_csv("data/Cost_of_Residual_Fuel_Oil_Delivered_to_Electric_Generating_Plants_Annual.csv")

data.head()
```

```python
China = data[data.Country == 'China']
US = data[data.Country == 'United States']
Germany = data[data.Country == 'Germany']
India = data[data.Country == 'India']
Japan = data[data.Country == 'Japan']
```

```python
plt.plot(US.Year, US.Solar_PV_Consumption) 
plt.plot(China.Year, China.Solar_PV_Consumption)
plt.plot(Germany.Year, Germany.Solar_PV_Consumption)
plt.plot(India.Year, India.Solar_PV_Consumption)
plt.plot(Japan.Year, Japan.Solar_PV_Consumption)
key = plt.legend(labels = ['US', 'China', 'Germany','India','Japan'], 
                 title= "Country", fontsize = 15)
rcParams['figure.figsize'] = 15, 15
key.legendHandles[0]._sizes = [40]
key.legendHandles[1]._sizes = [40]
key.legendHandles[2]._sizes = [40]
key.legendHandles[3]._sizes = [40]
plt.ylabel('Solar PV energy consumption, terawatt-hours per year')
plt.xlabel('Year')
plt.title('Solar PV consumption by country')
plt.savefig('Solar PV consumption by country.png',dpi=300)
```
